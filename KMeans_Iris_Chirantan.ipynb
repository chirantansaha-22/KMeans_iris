```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
data=iris.data
```

```python
labels=iris.target
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,7):
    km=KMeans(n_clusters=i,init='k-means++',random_state=0,n_init=7)
    km.fit(data)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, 7), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.xticks(np.arange(1, 7, 1))  # Set x-axis ticks to integers
plt.grid(True)
plt.show()
```

```python
#3 seems to be a good choice for number of clusters

kmeans=KMeans(n_clusters=3,init='k-means++',random_state=0,n_init=10)
y_means=kmeans.fit_predict(data)
centroids=kmeans.cluster_centers_
```

```python
from sklearn.metrics import accuracy_score,silhouette_score
```

```python
silhouette_score(data,y_means)
```
